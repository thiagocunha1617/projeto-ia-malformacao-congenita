```python
def delete_rows_with_values(df, columns, values):
    # Verificar se as listas de colunas e valores têm o mesmo tamanho
    if len(columns) != len(values):
        raise ValueError("O número de colunas deve ser igual ao número de valores.")

    # Iterar sobre as colunas e valores e deletar as linhas correspondentes
    for column, value in zip(columns, values):
        df = df[df[column] != value]
    
    # Retornar o DataFrame atualizado
    return df
```

```python
# Exemplo de uso da função para deletar linhas com valores indesejados em diferentes colunas
columns_to_exclude = data.columns
values_to_exclude = data.columns

data = delete_rows_with_values(data, columns_to_exclude, values_to_exclude)
```

```python
import numpy as np
from sklearn.metrics import classification_report
from tensorflow import keras

def ANN(X_train, y_train, X_test, y_test, loss, weights):
    model = keras.Sequential([
        keras.layers.Dense(11, input_dim=11, activation='relu'),
        keras.layers.Dense(15, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])
    
    if weights == -1:
        model.fit(X_train, y_train, epochs=100)
    else:
        model.fit(X_train, y_train, epochs=100, class_weight=weights)
    
    print(model.evaluate(X_test, y_test))
    
    y_preds = model.predict(X_test)
    y_preds = np.round(y_preds)
    
    print("Classification Report: \n", classification_report(y_test, y_preds))

    # Calcular a matriz de confusão
    cm = confusion_matrix(y_test, y_preds)
    print(cm)

    
    return y_preds
```

```python
from imblearn.under_sampling import RandomUnderSampler

# Criar o objeto RandomUnderSampler
rus = RandomUnderSampler(random_state=42)

# Aplicar o Random Undersampling nos dados
X_resampled, y_resampled = rus.fit_resample(X, y)

print(y_resampled.value_counts())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=15, stratify=y_resampled)
```

```python
import xgboost as xgb
from sklearn.metrics import confusion_matrix

# Treinar o modelo XGBoost
model = xgb.XGBClassifier()
model.fit(X_resampled, y_resampled)

# Fazer as predições
y_pred = model.predict(X_test)

print("Classification Report: \n", classification_report(y_test, y_pred))

# Calcular a matriz de confusão
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
