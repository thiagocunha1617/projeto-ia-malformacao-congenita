```python
# Creating the df_res dataframe with resample data
df_res = pd.concat([pd.DataFrame(X_res), pd.DataFrame(y_res)], axis=1)
```

```python
X = pd.DataFrame(df_res.drop(['IDANOMAL'], axis=1))
y = df_res.IDANOMAL
```

```python
# Train/Test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Importing the libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline

# Figure size
plt.figure(figsize=(12,8))

# Creating a list with all models
classifiers = [
    KNeighborsClassifier(2),
    GaussianNB(),
    LogisticRegression(),
    DecisionTreeClassifier(),
    RandomForestClassifier()]

# Creating a for loop to run the pipeline
for clf in classifiers:
    # fitting the model
    clf.fit(X_train, y_train)
    # models name
    name = clf.__class__.__name__
    print("="*30)
    print(name)
    # Priting the results
    print('****Results****')
    # Making the predictions
    y_pred = clf.predict(X_test)
    # Metrics
    print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
    print("Precision:", metrics.precision_score(y_test, y_pred))
    print("Recall:", metrics.recall_score(y_test, y_pred))
    print("F1-Score:", metrics.f1_score(y_test, y_pred))
    
    # Plotting the ROC curve
    y_pred_proba = clf.predict_proba(X_test)[::,1]
    fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
    auc = metrics.roc_auc_score(y_test, y_pred_proba)
    plt.plot(fpr,tpr,label=name+", auc="+str(auc))
    plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
    plt.legend(loc=4)
```
