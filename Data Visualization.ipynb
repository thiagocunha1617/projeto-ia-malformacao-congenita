```python
## missing percentage
1-(np.sum(df['congenital_anomaly'] == 2.0) + np.sum(df['congenital_anomaly'] == 1.0))/len(df['congenital_anomaly'])
```

```python
def get_missing_pct(file_name, state_name):
    if state_name == None:
        df = pd.read_csv(file_name)
        return 1-(np.sum(df['congenital_anomaly'] == 2.0) + np.sum(df['congenital_anomaly'] == 1.0))/len(df['congenital_anomaly'])
    else:
        df = pd.read_csv(file_name)
        df = df[df['Area']==state_name]
        return 1-(np.sum(df['congenital_anomaly'] == 2.0) + np.sum(df['congenital_anomaly'] == 1.0))/len(df['congenital_anomaly'])
```

```python
for i in range(7):
    file_name = "combined_data/LB" + str(2015+i) + ".csv"
    missing_pct = get_missing_pct(file_name, state_name=None)
    print(r"The missing percentage of year {} is {}".format(2015+i, missing_pct))
```

```python
locations = ["AC", "AL", "AP", "AM", "BA", "CE", "DF", "ES", "GO", "MA", "MT", "MS", "MG", "PA", "PB", 
              "PR", "PE", "PI", "RJ", "RN", "RS", "RO", "RR", "SC", "SP", "SE", "TO"]

years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
```

```python
## missingness by states
for state in locations:
    file_name = "combined_data/LB2020.csv"
    missing_pct = get_missing_pct(file_name, state)
    print(r"The missing percentage state {} is {}".format(state, missing_pct))
```

```python
df2020 = pd.read_csv("combined_data/LB2020.csv")
prevalence = []
for state in locations:
    sub_df = df2020[df2020['Area'] == state]
    prevalence.append(np.sum(sub_df['congenital_anomaly'] == 1.0)/sub_df.shape[0])

p_df = d = {'state':locations,'ca_preva':prevalence}    
prevalence_df = pd.DataFrame(p_df)
```

## Scrape Wiki table of GDP per capita cross Brazil states

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
wiki_url = 'https://en.wikipedia.org/wiki/Federative_units_of_Brazil'
response = requests.get(wiki_url)
soup = BeautifulSoup(response.text,'html.parser')
table = soup.find('table',{'class':'wikitable sortable'})
```

```python
gdp_df=pd.read_html(str(table))
# convert list to dataframe
gdp_df=pd.DataFrame(gdp_df[0])
gdp_df.head()
```

```python
gdp_df.columns, prevalence_df.columns
```

```python
gdp_prevalence = pd.merge(gdp_df, prevalence_df, how = 'inner', left_on = 'Code', right_on = 'state')
```

```python
gdp_prevalence.head()
```

```python
# gdp_prevalence['HDI(2022)[23]']
```

```python
plt.scatter(gdp_prevalence['HDI(2022)[23]'], gdp_prevalence['ca_preva'], alpha = 0.7)
m, b = np.polyfit(gdp_prevalence['HDI(2022)[23]'], gdp_prevalence['ca_preva'], 1)
plt.plot(gdp_prevalence['HDI(2022)[23]'], m*gdp_prevalence['HDI(2022)[23]']+b, color = "red")
plt.xlabel("Human Development Index")
plt.ylabel("Cogenital Malformation Prevalence")
plt.title("HDI vs Cogenital Malformation Prevalence in 2020")
```

```python
# gdp per capita vs congenital prevalence
plt.scatter(np.log(gdp_prevalence['GDP (R$millions, 2022)[22]']/gdp_prevalence['Population(2022)[21]']), gdp_prevalence['ca_preva'])
m, b = np.polyfit(np.log(gdp_prevalence['GDP (R$millions, 2022)[22]']/gdp_prevalence['Population(2022)[21]']), gdp_prevalence['ca_preva'], 1)
plt.plot(np.log(gdp_prevalence['GDP (R$millions, 2022)[22]']/gdp_prevalence['Population(2022)[21]']), m*np.log(gdp_prevalence['GDP (R$millions, 2022)[22]']/gdp_prevalence['Population(2022)[21]'])+b, color = "red")
plt.xlabel("GDP per capita in log scale")
plt.ylabel("Cogenital Malformation Prevalence")
plt.title("GDP per capita vs Cogenital Malformation Prevalence in 2020")
```
